```python
liste_gradcam = dict()
root_path = "./result/train_layercam/"
results = os.listdir(root_path)
for r in results:
    df = pd.read_csv(root_path+r)
    print(root_path+r)
    
    if "random" in r:
        df2 = df.mean(axis = 0)*2
    else:
        df2 = df.mean(axis = 0)*2+0.1
    
    model_name = r.split(".csv")[0].rsplit("_",1)[0]
    
    if model_name in liste_gradcam:
        liste_gradcam[model_name] += df2.values[1:]
    else:
        liste_gradcam[model_name] = df2.values[1:] 
    print(liste_gradcam[model_name])
    
    print(df2)
    print("-"*20)
```
